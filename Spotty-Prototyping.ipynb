```python
%matplotlib inline
import os
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
import skimage.feature as skimfeat
```

```python
# Specify a big font size
fontsize = 90

# Font paramters
scale = 95
aspect = 0.793
descent = 0.29
border = 0.2

# Make a figure large enough for exactly 1 character
figheight = (1+2*border)*fontsize*(1+descent)/scale
figwidth =  9

# Create the figure
fig = plt.figure(figsize=(figwidth,figheight))
ax = fig.add_axes((0,0,1,1))
ax.text(0.01*figwidth, (border+descent)/(1+2*border+descent), "A message!", fontsize=fontsize)
#ax.text(border, (border+descent)/(1+2*border+descent), "Happy Birthday, Katy!", family="monospace", fontsize=fontsize)

# Remove the axes
#ax.set_frame_on(False)
#ax.set_xticks([])
#ax.set_yticks([])
#plt.axis('off')

# Renderer
fig.canvas.draw()

# Convert to a numpy array
w,h = fig.canvas.get_width_height()
im = np.fromstring( fig.canvas.tostring_rgb(), dtype=np.uint8 )
im.shape = (h,w,3)

# Convert to black and white
immes = np.mean(im, axis=2)

# See what it looks like
plt.figure(figsize=(figwidth,figheight))
plt.imshow(immes, cmap='bone')
plt.axis("off");

```

```python
rad = 0.02*figheight

# Generate a random dot
col = np.array([0.1,0.2,0.6])
cen = (0.2,0.2)

fig = plt.figure(figsize=(figwidth,figheight))
ax = fig.add_axes((0,0,1,1))
ax.set_xlim((0,figwidth))
ax.set_ylim((0,figheight))

circle_artist = plt.Circle(cen,rad,color=col,fill=True)
ax.add_artist(circle_artist)

fig.show()
```
